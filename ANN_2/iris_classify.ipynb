# IRIS Classification using ANN

```python
#importing dataset from sklearn.org
from sklearn.datasets import load_iris

iris = load_iris() 
```

```python
#Displaying the column names of the dataset
dir(iris)
```

```python
#Displaying feature names in iris
iris.feature_names
```

```python
#Displaying data
iris.data
```

```python
#Displaying target column names
iris.target_names
```

```python
iris.target #0-setosa 1-versicolor ,2-virginica
```

```python
#Splitting the dataset into training and testing
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(iris.data,iris.target,test_size=0.2,random_state=1)

print(f"X_trainShape:{X_train.shape}\nX_testShape:{X_test.shape}\nY_trainShape:{y_train.shape}\nY_testShape:{y_test.shape}")
```

```python
#Creating ANN model
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()

model.add(Dense(16,input_shape=(4,))) #input layer
model.add(Dense(16,activation ='relu')) # hiddenlayer 1
model.add(Dense(16,activation='relu')) # hiddenlayer 2
model.add(Dense(3,activation='softmax')) #output layer

```

```python
#compling the model
model.compile(loss = 'sparse_categorical_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
```

```python
#Training the model
hist = model.fit(X_train,y_train,batch_size=16,epochs = 300 ,validation_data = (X_test,y_test),verbose=2)
hist
```

```python
#Evaluting the model
val_loss , val_acc = model.evaluate(X_test,y_test)
print(f"Val_Loss:{val_loss}\nVal_Accuracy:{val_acc}")
```

```python
#Making prediction on new data
import numpy as np
flower_input = np.array([[5.1,3.5,1.4,0.2]])
result = model.predict(flower_input)
if (np.argmax(result)) == 0:
    print("Setosa [0]")
elif (np.argmax(result)) == 1:
    print("versicolor [1]")
elif (np.argmax(result)) == 2:
    print("Virginica [2]")
else:
    pass
```

```python
#Visualization of accuracy on each epoch
import matplotlib.pyplot as plt

plt.plot(hist.history['accuracy'],color='lime')
plt.plot(hist.history['val_accuracy'],color='blue')

plt.title("Model Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend(['train','test'],loc='lower right')
plt.style.use('dark_background')
plt.show()
```

```python
#visualization on loss function at each epoch
plt.plot(hist.history['loss'],color='lime')
plt.plot(hist.history['val_loss'],color='blue')

plt.title("Model Loss")
plt.xlabel('Epochs')
plt.ylabel("Loss")
plt.legend(['train','test'],loc='upper right')
plt.style.use('dark_background')
plt.show()
```

```python
#Confusion Matrix plot

from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay

y_pred_prob = model.predict(X_test)
y_pred = np.argmax(y_pred_prob,axis=1)
cm = confusion_matrix(y_test,y_pred)
disp = ConfusionMatrixDisplay(cm,display_labels=['Setosa[0]','versicolor [1]','Virginica [2]'])

disp.plot(cmap='inferno')
plt.show()
```
